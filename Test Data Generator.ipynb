```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import numpy as np
import load_data
import pathlib
```

```python
dim_old = 28
dim_new = None
dat_path = pathlib.Path('object_files/').glob('*.npz')
d = {"lollipop":0, "canoe":1, "penguin":2, "eyeglasses":3, "apple":4, "moon":5, "cup":6}
train, train_labels, valid, valid_labels, test, test_labels = load_data.load_for_cnn(dat_path, 
                                                                                     dim_old, d, (0.7, 0.25), 
                                                                                     channel_axes=0, 
                                                                                     dim_new=dim_new)
```

```python
batch_size = 64
gen = load_data.ImageDataGenerator(train, train_labels, batch_size, 1)
train_datagen = gen.flow()
```
